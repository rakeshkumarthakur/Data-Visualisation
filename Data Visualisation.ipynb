```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\Rakesh Thakur\Downloads\2W Data.xlsx")
```

```python
df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)

df.describe()
```

```python
# Bar Graph
plt.figure(figsize=(20,5))
sns.barplot(x="Year", y="Total Sales", data=df, color="red")
plt.title("Total Sales Over Years")
plt.ylabel("Total Sales")
plt.xlabel("Year")
plt.show()
```

```python
# Line plot
plt.figure(figsize=(20,5))
sns.lineplot(x="Year", y="Total Sales", data=df, marker="o")
plt.title("Total Sales Over Years")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Bar Chart
plt.figure(figsize=(20,5))
sns.barplot(x="Year", y="Total Sales", data=df, palette="Blues_d")
plt.title("Total Sales by Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Multiple Line Plot (Motorcycles, Scooters, Mopeds)
plt.figure(figsize=(20,5))
sns.lineplot(x="Year", y="Motorcycles", data=df, marker="o", label="Motorcycles")
sns.lineplot(x="Year", y="Scooters", data=df, marker="o", label="Scooters")
sns.lineplot(x="Year", y="Mopeds", data=df, marker="o", label="Mopeds")
plt.title("Vehicle Sales Over Years")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#Stacked Bar Chart (Motorcycles, Scooters, Mopeds per Year)
plt.figure(figsize=(8,5))
df.set_index("Year")[["Motorcycles","Scooters","Mopeds"]].plot(
    kind="bar", stacked=True, figsize=(8,5), colormap="tab20"
)
plt.title("Stacked Sales by Vehicle Type")
plt.ylabel("Sales")
plt.show()
```

```python
#Pie Chart
year = "FY2025"
values = df.loc[df["Year"]==year, ["Motorcycles","Scooters","Mopeds"]].values.flatten()
labels = ["Motorcycles","Scooters","Mopeds"]

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title(f"Vehicle Sales Distribution in {year}")
plt.show()
```

```python
#Heatmap (Correlation between Sales Types)
plt.figure(figsize=(6,4))
sns.heatmap(df.drop(columns="Year").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Sales Types")
plt.show()
```

```python
sns.pairplot(df)
```

```python
#Area Chart (Total Sales Trend)
plt.figure(figsize=(20,5))
plt.fill_between(df["Year"], df["Total Sales"], color="skyblue", alpha=0.4)
plt.plot(df["Year"], df["Total Sales"], marker="o", color="blue")
plt.title("Total Sales Area Chart")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Grouped Bar Chart (Motorcycles vs Scooters vs Mopeds per Year)
plt.figure(figsize=(20,5))
df_melt = df.melt(id_vars="Year", value_vars=["Motorcycles","Scooters","Mopeds"],
                  var_name="Vehicle", value_name="Sales")
sns.barplot(x="Year", y="Sales", hue="Vehicle", data=df_melt)
plt.title("Vehicle Sales Comparison by Year")
plt.ylabel("Sales")
plt.show()
```

```python
#Box Plot (Distribution of Vehicle Sales)
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["Motorcycles","Scooters","Mopeds"]])
plt.title("Distribution of Vehicle Sales")
plt.ylabel("Sales")
plt.show()
```

```python
#Violin Plot (Sales Distribution Shape)
plt.figure(figsize=(8,5))
sns.violinplot(data=df[["Motorcycles","Scooters","Mopeds"]])
plt.title("Sales Distribution (Violin Plot)")
plt.ylabel("Sales")
plt.show()
```
